```python
count_df=pd.DataFrame(tot_df.groupby('Book-Title')['Book-Rating'].count().sort_values(ascending=False)) 
count_df.head(20)
```

```python
count_df=count_df.merge(tot_df.groupby('Book-Title')['Book-Rating'].mean(),on='Book-Title')
```

```python
count_df.reset_index()
count_df.rename(columns={'Book-Rating_x':'Rating_count','Book-Rating_y':'Mean'},inplace=True)
count_df.head()
```

```python
rating_cutoff=count_df['Rating_count'].quantile(.97)
print(rating_cutoff)
mean_cutoff=count_df['Mean'].mean()
print(mean_cutoff)
```

```python
l=count_df.loc[(count_df['Rating_count']>=rating_cutoff) & (count_df['Mean']>=mean_cutoff)].index
```

```python
print(l)
```

```python
count_df.index
```

```python
tot_df=tot_df.loc[tot_df['Book-Title'].isin(l)]
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
pivot=tot_df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot.fillna(0,inplace=True)
```

```python
pivot.sample(20)
```

```python
from scipy.sparse import csr_matrix
```

```python
sparse=csr_matrix(pivot)
```

```python
nn=NearestNeighbors(metric='cosine',algorithm='brute')
nn.fit(sparse)
```

```python
distances, indices=nn.kneighbors(pivot.iloc[1000,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances[0]
```

```python
print(pivot.index[1000])
index=pivot.index 
for i in range(len(distances[0])):
    print(index[indices[0][i]])
```
